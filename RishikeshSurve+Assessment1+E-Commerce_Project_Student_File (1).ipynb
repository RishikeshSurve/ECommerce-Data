## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed. (6 marks)

#### Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total Revenue. 

<font color='blue'>
    <font size = 5>
    Now you need to find the revenue for each retailer (Brand)
    </font>

```python
#def charges for each catergory as per the problem set
charges = {
    'greater_than_600' : 0.25,
    'greater_than_350_or_less_than_equal_to_600' : 0.15,
    'greater_than_100_or_less_than_equal_to_350' : 0.10,
    'Other' : 0.05
}

#def func to calculate revenue for each category
def cal_revenue(discounted_price, percentage):
    return discounted_price*percentage

#revenue for each category mentioned above
revenue_greater_than_600 = cal_revenue(ECom_Data_2[ECom_Data_2['discounted_price'] > 600]['discounted_price'].sum(), charges['greater_than_600'])
revenue_greater_than_350_or_less_than_equal_to_600 = cal_revenue(ECom_Data_2[ECom_Data_2['discounted_price'] <= 600]['discounted_price'].sum(), charges['greater_than_350_or_less_than_equal_to_600'])
revenue_greater_than_100_or_less_than_equal_to_350 = cal_revenue(ECom_Data_2[ECom_Data_2['discounted_price'] <= 350]['discounted_price'].sum(), charges['greater_than_100_or_less_than_equal_to_350'])
revenue_Other = cal_revenue(ECom_Data_2[ECom_Data_2['discounted_price'] <= 100]['discounted_price'].sum(), charges['Other'])

#sum
Total_Revenue = revenue_greater_than_600 + revenue_greater_than_350_or_less_than_equal_to_600 + revenue_greater_than_100_or_less_than_equal_to_350 + revenue_Other

#Total_Revenue
print("Total Revenue generated by the E-Commerce company over all orders placed is :", Total_Revenue)
```

<font color = 'green'>
    
### 1. Initially we defined a variable called "charges" in which we defined the amount of percentage the company was charging.   
### 2. Then we created a def function called "cal_revenue" in which we defined how the revenue should be calculated using the "discounted_price" and "percentage". This helps us to directly get the calculation (multiplication) of "charges" and "discounted price".
### 3. We calculated the revenue for each category mentioned under the variable "charges". out of which we got the "Total_Revenue" by doing the sum as mentioned above.
### 4. Print the "Total_Revenue" for the output

## Q 9.Compare performance regionwise

## a)	Draw a lineplot for the monthly Revenue of E-Commerce Company for each region separately.  (4 marks) 

## b)	Identify the best and the worst performing months for each region. (2 marks)

#### Note: Only those days with actual orders(Order_Date) placed are present in the dataset. Assuming there were no orders on other days. Also, show Month and Year on x-axis.

### Ans 9 a)

```python
# Convert 'Order_Date' column to datetime format
brand_info['Order_Date'] = pd.to_datetime(brand_info['Order_Date'])

# Extract month from the 'Order_Date' column
brand_info['Month'] = brand_info['Order_Date'].dt.month

#Group by 'Region' and 'Month' and sum up the revenue
monthly_revenue_by_region = brand_info.groupby(['Region', 'Month'])['discounted_price'].sum().reset_index()

# Plot line plots for each region separately
for region in monthly_revenue_by_region['Region'].unique():
    region_data = monthly_revenue_by_region[monthly_revenue_by_region['Region'] == region]
    plt.plot(region_data['Month'], region_data['discounted_price'], label=region)

# Set plot title and labels
plt.title('Monthly Revenue of E-Commerce Company by Region')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.legend(title='Region')
plt.grid(True)

# Show plot
plt.show()
```

### Ans 9 b)

```python
#Group the data by Region and Month, and calculate total revenue
monthly_revenue = brand_info.groupby(['Region', brand_info['Order_Date'].dt.month])['discounted_price'].sum().reset_index()

#Find the best and worst performing months for each region
best_months = monthly_revenue.loc[monthly_revenue.groupby('Region')['discounted_price'].idxmax()]
worst_months = monthly_revenue.loc[monthly_revenue.groupby('Region')['discounted_price'].idxmin()]

print("Best Performing Months:")
print(best_months)

print("\nWorst Performing Months:")
print(worst_months)
```
